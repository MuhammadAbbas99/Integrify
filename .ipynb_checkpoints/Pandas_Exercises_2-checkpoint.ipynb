### Step 4. How many products cost more than $10.00?

```python
#chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x.replace('$','')))
price_10 = chipo[chipo['item_price']>10]
unique=price_10['item_price'].count()
products = unique.nunique()
#chipo[chipo['item_price']]
```
